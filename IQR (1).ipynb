```python
import pandas as pd
import numpy as np
import seaborn as sns
df=pd.read_csv('bhp (1).csv')
df
```

```python
df.info()
```

```python
sns.boxplot(df.price_per_sqft)
```

```python
#Calculating Q1 and Q3
Q1=np.percentile(df.price_per_sqft,25)
Q3=np.percentile(df.price_per_sqft,75)
print("The Value of Q1 = ",Q1)
print('The value of Q3 = ',Q3)
```

```python
IQR=Q3-Q1
print("Interquartile range is: ", IQR)
```

```python
low_bound=Q1-1.5*IQR
up_bound=Q3+1.5*IQR
print('The lower bound value is : ',low_bound)
print('The upper bound value is : ',up_bound)
```

```python
df[(df.price_per_sqft<low_bound)|(df.price_per_sqft>up_bound)]
```

```python
no_outlier=df[(df.price_per_sqft<up_bound) & (df.price_per_sqft>low_bound)]
```

```python
no_outlier
```

```python
sns.boxplot(no_outlier.price_per_sqft)
```
